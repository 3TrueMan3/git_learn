```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('train.csv').sort_values(by='medv')   #Считываем файл и сортируем по возрастанию средней цены
```

```python
#Формируем значения по осям и строим график
x_data, y_data = df['medv'], df['dis']
plt.xlabel('medv')
plt.ylabel('dis')
plt.plot(x_data, y_data)
```

```python
#Получаем среднее по столбцу 'medv' и формируем новый DataFrame, затем записываем его в файл
df_rich = pd.DataFrame(df[df['medv'] > np.mean(np.array(df['medv']))])
df_rich.to_csv('rich_boston.csv')
```
